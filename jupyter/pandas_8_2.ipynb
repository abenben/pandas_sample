```python
% matplotlib inline
import pandas as pd
import numpy as np
```

# 8.2.1 plotメソッド

```python
np.random.seed(0)
val = np.random.randint(-5, 5, size=500).reshape(100, 5)
df = pd.DataFrame(val, columns=list('abcde')).cumsum()
df.head()
```

```python
df['a'].plot(kind='line')
```

```python
df.plot(kind='line')
```

```python
df['b'].plot(title='column b lint bot', legend=True)
```

```python
df['c'].plot(grid=True, xlim=(0, 20), ylim=(-30, 30))
```

```python
df['d'].plot(xticks=np.arange(0, 100, 5), grid=True)
```

```python
df['b'].plot(figsize=(3, 2))
```

```python
df[['a', 'b']].plot(subplots=True)
```
